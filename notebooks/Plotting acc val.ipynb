# Plotting Acc/ Val

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.style.use('ggplot')
logs_path = os.path.join(os.getcwd(), 'logs')
plots = [plot for plot in os.listdir(logs_path) if plot[-3:] == 'csv']
plots.sort()

train_acc = plots[0]
train_loss = plots[1]
val_acc = plots[2]
val_loss = plots[3]
```

```python
steps = list(pd.read_csv(os.path.join(logs_path, train_loss), usecols=[1]).Step)
train_acc = list(pd.read_csv(os.path.join(logs_path, train_acc), usecols=[2]).Value)
val_acc = list(pd.read_csv(os.path.join(logs_path, val_acc), usecols=[2]).Value)
train_loss = list(pd.read_csv(os.path.join(logs_path, train_loss), usecols=[2]).Value)
val_loss = list(pd.read_csv(os.path.join(logs_path, val_loss), usecols=[2]).Value)
```

```python
fig, ax = plt.subplots(2,1, sharex=True)

ax[0].plot(steps, train_acc, label =  'train')
ax[0].plot(steps, val_acc, label = 'val')
ax[0].set_title('Accuracy Plot')
ax[0].legend(loc="upper left")

ax[1].plot(steps, train_loss, label =  'train')
ax[1].plot(steps, val_loss, label = 'val')
ax[1].set_title('Loss Plot')
ax[1].legend(loc="upper right")
```
